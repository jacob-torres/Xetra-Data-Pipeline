```python
# Define expected data report
columns_report = [
    'isin', 'date', 'opening_price_eur', 'closing_price_eur',
    'minimum_price_eur', 'maximum_price_eur', 'daily_traded_volume', 'change_prev_closing_%'
]

data_report = [
    ['AT0000A0E9W5', '2021-04-17', 20.21, 18.27, 18.21, 21.34, 1088, 10.62],
    ['AT0000A0E9W5', '2021-04-18', 20.58, 19.27, 18.89, 21.14, 10286, 1.83],
    ['AT0000A0E9W5', '2021-04-19', 23.58, 24.22, 22.21, 25.01, 3586, 14.58]
]
df_report = pd.DataFrame(data_report, columns=columns_report)

print(df_report.shape)
df_report
```

```python
# Define transform function
def transform(data_frame):

    # Check for empty dataframe
    if data_frame.empty:
        return data_frame

    # Select specific columns and drop all null values
    data_frame = data_frame.loc[:, columns_src]
    data_frame.dropna(inplace=True)

    # The opening_price column is created by sorting the data by Time,
    # then grouping it by ISIN and Date, selecting StartPrice,
    # and finally transforming the column to contain only the first date
    data_frame['opening_price_eur'] = (
        data_frame.sort_values(by=['Time'])
        .groupby(['ISIN', 'Date'])['StartPrice']
        .transform('first')
    )

    # The closing_price column is created by transforming the data
    # similarly to opening_price, but selecting for the last date instead
    data_frame['closing_price_eur'] = (
        data_frame.sort_values(by=['Time'])
        .groupby(['ISIN', 'Date'])['StartPrice']
        .transform('last')
    )

    # Rename columns
    data_frame.rename(columns={
        'ISIN': 'isin',
        'Date': 'date',
        'MinPrice': 'minimum_price_eur',
        'MaxPrice': 'maximum_price_eur',
        'TradedVolume': 'daily_traded_volume'
    }, inplace=True)

    # Data aggregation
    data_frame = (
        data_frame.groupby(['isin', 'date'], as_index=False)
        .agg({
            'opening_price_eur': 'min',
            'closing_price_eur': 'min',
            'minimum_price_eur': 'min',
            'maximum_price_eur': 'max',
            'daily_traded_volume': 'sum'
    })
    )

    # The prev_closing_price column is created
    # by sorting the data by Date, and then grouping it by ISIN
    # and selecting for closing price of the previous date
    data_frame['change_prev_closing_%'] = (
        data_frame.sort_values(by=['date'])
        .groupby(['isin'])
        ['opening_price_eur'].shift(1)
    )

    # Calculate the percentage of change in the closing price since the last date
    data_frame['change_prev_closing_%'] = (
        (data_frame['opening_price_eur'] -
            data_frame['change_prev_closing_%']
        ) / data_frame['change_prev_closing_%'] * 100
    )

    # Round all float values to 2 decimals
    data_frame = data_frame.round(decimals=2)

    # Filter the dataframe by date
    data_frame = data_frame[
        data_frame.date >= extract_date
        ].reset_index(drop=True)

    return data_frame
```

```python
# Check the resulting dataframe after processing through the transform function
df_result = transform(df_src.loc[1:8])

print(df_result.shape)
df_result
```
